```python
import math

saving = {
    "goal" : 0,
    "salary" :  0,
    "financial_obligations" : []
}

financial_sum = 0

def add_saving_plan():
    global financial_sum
    while True:
        try:
            income = float(input("\nEnter your income: "))
            if income <= 0:  
                print("Income cannot be negative or 0.")
                continue
            break
        except:
            print("Please enter a valid number for income.")
    
    while True:    
        try:
            goal = float(input("\nEnter your goal: "))
            if goal <= 0:  
                print("Goal cannot be negative or 0.")
                continue
            break
        except:
            print("Please enter a valid number for goal.")
    
    financial_obligations = []
    while True:
        fo = input("\nEnter a financial obligation (or 'done' to finish): ")
        if fo.lower() == 'done':
            print()
            break
        try:
            if float(fo) > income or float(fo) < 0:
                print("financial obligations must be less than income and greater than 0")
                continue
            financial_obligations.append(float(fo))  
        except:
            print("Please enter a valid number for financial obligations.")

    saving["income"] = income
    saving["goal"] = goal
    saving["financial_obligations"] = financial_obligations

    for i in financial_obligations:
        if financial_sum < income:
            financial_sum += i

def display_plan(x):
    if not x or not any(x.values()):
        print("You don't have a plan! \n")
    else:
        print("\nPlan Info:")
        print("\n Income :" , x["income"])
        print(" Goal :" , x["goal"])
        print(" Financial obligations :", financial_sum)
        print(" Net save: " , x["income"] - financial_sum )
        
        
        calc_months = lambda x, sum_obligations: math.ceil(x["goal"] / (x["income"] - sum_obligations))
        months = calc_months(x, financial_sum)  # Using the lambda function
        print(" Months: ", months)
        print()

def delete_plan():
    global saving
    if not saving or not any(saving.values()):
        print("You don't have a plan to delete it! \n")
    else:
        saving = {
        "goal" : 0,
        "salary" :  0,
        "financial_obligations" : []
        }
        print("Plan Deleted! \n")

while True:
    print("1. Add new plan")
    print("2. Display plan")
    print("3. Delete plan")
    print("4. Exit \n")

    x = input("Enter number: ")

    if(x == "1"):
        if not saving or not any(saving.values()):
            add_saving_plan()
        else:
            print("You are already added plan! \n")
    elif (x == "2"):
        display_plan(saving)

    elif (x == "3"):
        delete_plan()
        
    elif (x == "4"):
        print("Goodbye")
        break
    else:
        print("Please enter valid number from the choices \n")
```
